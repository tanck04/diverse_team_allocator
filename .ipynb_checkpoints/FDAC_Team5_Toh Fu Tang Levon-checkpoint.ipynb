To read a CSV file and convert it into a list of dictionaries, we follow these steps:
First, we define the read_csv_to_dict function, which takes one par1. ameter:
file_path: The path to the CSV file.
Next, we open the file in read mode: We read all lines from the file and store them in a list called lines.
We then strip newline characters and split the lines into a list of lists. This list is stored in data.
The first row of data is the header, which contains the column names. We store this row in headers.
We create an empty list called result to hold the resulting dictionaries.
We iterate over the remaining 1. rows in data:
For each row, we create a dictionary where the keys are the column names (from headers) and the values are the corresponding values2.  from the row.
We append this dictionary to result.
Finally, the function returns result, which is a list of dictionaries representing the rows of the CSV file.


```python
# Create a function to read CSV file as lists
def read_csv_to_list(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        # Initialize an empty list to hold the rows
        csv_reader = []
        
        # Read the file line by line
        for line in file:
            # Strip newline characters and split by comma
            row = line.strip().split(',')
            # Append the row to csv_reader
            csv_reader.append(row)
    
    return csv_reader

# Create a function to read CSV file as dictionaries, with column headers as keys, content in each rows as values
def read_csv_to_dict(file_path):
    """Reads a CSV file and returns a list of dictionaries, simulating csv.DictReader."""
    with open(file_path, 'r') as file:
        # Read all lines from the file
        lines = file.readlines()

    # Strip newline characters and split the lines into a list of lists
    data = [line.strip().split(',') for line in lines]

    # The first row is the header
    headers = data[0]

    # Create a list to hold the resulting dictionaries
    result = []

    # Iterate over the remaining rows
    for row in data[1:]:
        # Create a dictionary for each row
        row_dict = {headers[i]: row[i] for i in range(len(headers))}
        result.append(row_dict)

    return result
```

__6. Gini coefficient gender__
To ensure gender diversity, we are creating a coefficient for gender which will be utilized in the calculation of the inequality score which quantifies the inequality. 
The parameters provided are males and females. 
We put a safeguard against dividing by zero and store the number of total students in ‘total_people’. 
Then, we count the number of students in each gender category and store both in ‘n’. 
To calculate the gini coefficient,
 First we calculate the absolute value of the difference in the number of  male and females, i.e, we store the excess number of students in either category in ‘numerator’. 
 We do this as this value helps us account for the inequality or equality in the number of students in each gender.
Then we divide the numerator by two times the square of ‘total_people’ and store this result as ‘gini _coefficient’.    
We do this to create a coefficient which takes in account the difference of males and females with respect to the number of total people. 
Now, we create a ‘penalty_factor’ which is two times the quartic of ‘total_people’. 
We created this to account for how inequality generally decreases as the number of people increases.
 Lastly we add the penalty factor to the previous value of ‘gini_coefficient’ to get the final “gini_coefficient” which has taken gender into account for diversity. 
This a function which accounts for the gender diversity quota and adds to the ‘gini_coefficient’


```python
def calculate_gini_coefficient_gender(males, females):
    total_people = males + females
    
    if total_people == 0:
        return 0.0  # To avoid division by zero
    
    gender_counts = [males, females]
    n = len(gender_counts)
    
    # Calculate Gini coefficient
    numerator = 0
    for i in range(n):
        for j in range(n):
            numerator += abs(gender_counts[i] - gender_counts[j])
    
    gini_coefficient = numerator / (2 * total_people * total_people)
    penalty_factor=2*total_people**4
    gini_coefficient=gini_coefficient+penalty_factor

    return gini_coefficient
```

__7. Gini coefficient school__
To ensure diversity in the schools of each student, we are creating a coefficient for school which will be utilized in the calculation of the inequality score which quantifies the inequality. 
First we extract the number of students in each school from the given parameter of ‘school_dict’a\ and store them in ‘student_counts’.
Then we store the number of schools in ‘n’.
Next,we put a safeguard against dividing by zero by creating an if function which returns 0.0 if the number of schools is 0.
Then, we calculate the mean number of students by dividing the sum of values in ‘student_count’ by the number of schools.
To calculate the gini coefficient,
 First we initialize the numerator to 0 and then loop through each pair of schools (i,j) and add the absolute difference between their student counts, i.e, the number of students in each school to the numerator.. 
 We do this as this value helps us account for the inequality or equality in the number of students in each school with every other school.
Then we divide the numerator by two times the number of schools times the number of students and store this result as ‘gini _coefficient’.    
We do this to create a coefficient which takes in account the difference of students in different schools with respect to the number of total students and number of total schools.
Now, we create a ‘penalty_factor’ which is two times the square of the total number of schools. 
We created this to account for how inequality generally decreases as the number of schools increases.
 Lastly we add the penalty factor to the previous value of ‘gini_coefficient’ to get the final “gini_coefficient” which has taken school into account for diversity. 
This is a function which accounts for the school diversity quota.


```python
def calculate_gini_coefficient_school(school_dict):
    # Extract the number of students
    student_counts = list(school_dict.values())
    n = len(student_counts)
    
    if n == 0:
        return 0.0  # No schools
    
    # Calculate the mean number of students
    mean_students = sum(student_counts) / n

    # Calculate the Gini coefficient
    numerator = 0
    for i in range(n):
        for j in range(n):
            numerator += abs(student_counts[i] - student_counts[j])
    
    gini_coefficient = numerator / (2 * n * sum(student_counts))
    penalty_factor=2*n**2
    gini_coefficient=gini_coefficient+penalty_factor
    
    return gini_coefficient
```

```python
def calculate_gini_coefficient_school_no_penalty(school_dict):
    # Extract the number of students
    student_counts = list(school_dict.values())
    n = len(student_counts)
    
    if n == 0:
        return 0.0  # No schools
    
    # Calculate the mean number of students
    mean_students = sum(student_counts) / n

    # Calculate the Gini coefficient
    numerator = 0
    for i in range(n):
        for j in range(n):
            numerator += abs(student_counts[i] - student_counts[j])
    
    gini_coefficient = numerator / (2 * n * sum(student_counts))
    
    return gini_coefficient
```

__9. Gini coefficient cgpa__
To ensure diversity in the schools of each student, we are creating a coefficient for school which will be utilized in the calculation of the inequality score which quantifies the inequality. 
First we extract the number of CGPAs- this is the number of students-from cgpas and store it in ‘n’.
Next,we put a safeguard against dividing by zero by creating an if function which returns 0.0 if the number of students is 0.
Then, we convert all elements in cpgas i.e, all CGPA values into floats. This is for accuracy and precision in calculation.
We calculate the average CGPA by dividing the sum of values in ‘cgpas’ by the number of students.
To calculate the gini coefficient,
 First we initialize the numerator to 0 and then loop through each pair of students (i,j) and add the absolute difference between their CGPAs, to the numerator.
 We do this as this value helps us account for the inequality or equality in the CGPAs of each student with every other student. 
Then we divide the numerator by two times the number of students times the sum of the values in ‘cgpas’ and store this result as ‘gini _coefficient’.    
We do this to create a coefficient which takes in account the difference of cgpas with respect to the number of total students and the sum of all CGPA values.
Now, we create a ‘penalty_factor’ which is two times the square of the total number of students. 
We created this to account for how inequality generally decreases as the number of students increases.
 Lastly we add the penalty factor to the previous value of ‘gini_coefficient’ to get the final “gini_coefficient” which has taken cgpa into account for diversity. 
This is a function which accounts for the school diversity quota.


```python
def calculate_gini_coefficient_cgpa(cgpas):
    n = len(cgpas)
    
    if n == 0:
        return 0.0  # No CGPAs
    cgpas=[float(ele)for ele in cgpas]
    # Calculate the mean CGPA
    mean_cgpa = sum(cgpas) / n

    # Calculate the Gini coefficient
    numerator = 0
    for i in range(n):
        for j in range(n):
            numerator += abs(cgpas[i] - cgpas[j])
    
    gini_coefficient = numerator / (2 * n * sum(cgpas))
    penalty_factor=2*n**2
    gini_coefficient=gini_coefficient+penalty_factor

    return gini_coefficient
```

__10. Calculate stats__
To calculate the overall inequality score for a list of students, we follow these steps:
First, we initialize variables M and F to 0 to count the number of male and female students, respectively. We also initialize an empty dictionary schools to count the number of students in each school, and an empty list CGPA to store the CGPAs of all students.
Next, we iterate through each student in list_of_students:
If the student’s gender is ‘Male’, we increment M by 1. Otherwise, we increment F by 1.
We then check if the student’s school is already a key in the schools dictionary. If it is, we increment the count for that school by 1. If it is not, we add the school to the dictionary with a count of 1.
We also append the student’s CGPA to the CGPA list.
After processing all students, we calculate three Gini coefficients:
G is the Gini coefficient for gender, calculated using the calculate_gini_coefficient_gender function and raised to the power of 3.
S is the Gini coefficient for schools, calculated using the calculate_gini_coefficient_school function and raised to the power of 3 * school_inequality.
C is the Gini coefficient for CGPAs, calculated using the calculate_gini_coefficient_cgpa function and raised to the power of 2.
Finally, we calculate the overall inequality score final as a weighted sum of G, S, and C, with weights of 0.4, 0.4, and 0.2, respectively.
The function then returns the final inequality score.

```python
def calculate_stats(list_of_students,school_inequality):
    M=0
    F=0
    schools={}
    CGPA=[]
    for student in list_of_students:
        if student['Gender']=='Male':
            M+=1
        else:
            F+=1
        if student['School'] in schools.keys():
            schools[student['School']]=schools[student['School']]+1
        else:
            schools[student['School']]=1
        CGPA.append(student['CGPA'])
        
    G=calculate_gini_coefficient_gender(M,F)**3
    S=calculate_gini_coefficient_school(schools)**(3*school_inequality)
    C=calculate_gini_coefficient_cgpa(CGPA)**2
    final=0.4*G+0.4*S+0.2*C
    return final
    
```

__11. Merge dictionaries__
To merge two dictionaries and create a new dictionary with values in alternating order, we follow these steps:
First, we create a list called values_list to hold the values from both dictionaries in an alternating order.
Next, we determine the maximum length of the two dictionaries using the max function. This ensures that our loop will run for the length of the longer dictionary.
Then, we iterate through the range of this maximum length. For each index i:
If i is a valid index for dict1 (i.e., i < len(dict1)), we append the value at the i-th key of dict1 to values_list.
Similarly, if i is a valid index for dict2 (i.e., i < len(dict2)), we append the value at the i-th key of dict2 to values_list.
After collecting all values in values_list, we convert this list into a new dictionary called new_dict. We use the enumerate function to assign a new index as the key for each value in the list.
Finally, we calculate the number of key-value pairs in new_dict and store this in num_pairs.
The function then returns new_dict and num_pairs.


```python
def merge_dictionaries(dict1, dict2):
    # Create a list to hold the values in alternating order
    values_list = []

    # Get the maximum length of the two dictionaries
    max_length = max(len(dict1), len(dict2))

    # Iterate through the range of maximum length
    for i in range(max_length):
        if i < len(dict1):  # Check if index is valid for dict1
            values_list.append(dict1[list(dict1.keys())[i]])
        
        if i < len(dict2):  # Check if index is valid for dict2
            values_list.append(dict2[list(dict2.keys())[i]])

    # Convert the list to a new dictionary with new index as the key
    new_dict = {index: value for index, value in enumerate(values_list)}

    # Get the number of key-value pairs
    num_pairs = len(new_dict)

    return new_dict, num_pairs
```

__12. Sorter__
To sort a dictionary of students by CGPA using bubble sort, we follow these steps:
First, we define the sorter function, which takes two parameters:
student_dict: A dictionary where keys are student IDs and values are dictionaries containing student information.
descending: A boolean parameter that determines the sort order. If True, the dictionary is sorted in descending order; otherwise, it is sorted in ascending order.
Next, we convert the dictionary items to a list of tuples for sorting. This list is stored in items, and n is set to the length of this list.
We then implement the bubble sort algorithm:
We use a nested loop to iterate through the list. The outer loop runs from 0 to n-1, and the inner loop runs from 0 to n-i-1.
For each pair of adjacent items, we compare their CGPA values based on the desired order:
If descending is False and the CGPA of the current item is greater than the CGPA of the next item, or if descending is True and the CGPA of the current item is less than the CGPA of the next item, we swap the two items.
This process ensures that the list is sorted by CGPA in the specified order.
After sorting, we convert the sorted list of tuples back into a dictionary called sorted_dict.
Finally, the function returns sorted_dict.


```python
# Create a function to sort students by CGPA in dictionary, bubble sort is employed
def sorter(student_dict, descending=False):
    """
    Sort a dictionary of students by CGPA using bubble sort.

    Parameters:
    student_dict (dict): A dictionary where keys are student IDs and values are dictionaries with student info.
    descending (bool): If True, sort in descending order; otherwise, sort in ascending order.

    Returns:
    dict: A new dictionary sorted by CGPA.
    """
    # Convert the dictionary items to a list of tuples for sorting
    items = list(student_dict.items())
    n = len(items)

    # Bubble sort algorithm
    for i in range(n):
        for j in range(0, n-i-1):
            # Compare CGPA values of adjacent items based on the desired order
            if (not descending and items[j][1]["CGPA"] > items[j + 1][1]["CGPA"]) or \
               (descending and items[j][1]["CGPA"] < items[j + 1][1]["CGPA"]):
                # Swap if the current item is out of order
                items[j], items[j + 1] = items[j + 1], items[j]

    # Convert the sorted list of tuples back into a dictionary
    sorted_dict = {key: value for key, value in items}
    return sorted_dict


#Guide to Use
'''
# Sort in ascending order
sorted_student_dict_ascending = sorter(student_dict)

# Sort in descending order
sorted_student_dict_descending = sorter(student_dict, descending=True)

'''
```

__13. Insert initial__
To insert two students initially into tutorial groups based on their batch, we use three parameters :
batch: The batch number of students.
records_indexed_dict_copy: A dictionary of student records indexed by their position.
tutorial_groupings_dict_copy: A dictionary of tutorial groupings.
First, we calculate the starting and ending indices for the batch:
starting_index is calculated as (batch - 1) * 50.
ending_index is calculated as batch * 50.
We then determine the tutorial group for the batch using the starting index:
We create a new dictionary batch_records_indexed_dict_copy containing only the records for the current batch:
We do this by filtering records_indexed_dict_copy to include only the records with keys between starting_index and ending_index.
We initialize an empty dictionary schools to count the number of students in each school:
We iterate through batch_records_indexed_dict_copy and update the schools dictionary with the count of students for each school.
We calculate the school inequality using the calculate_gini_coefficient_school_no_penalty function.
We create two dictionaries to separate male and female students:
males_batch_records_indexed_dict_copy contains records of male students.
females_batch_records_indexed_dict_copy contains records of female students.
We sort the male and female students by CGPA:
descending_males_batch_records_indexed_dict_copy is sorted in descending order.
ascending_females_batch_records_indexed_dict_copy is sorted in ascending order.
We then re-index these sorted dictionaries:
This ensures that the indices start from 0 and increment sequentially.
We insert the top 10 male and female students into the tutorial groups:
We iterate from 0 to 9 and append the corresponding male and female students to the tutorial group tut_grp.
We create dictionaries for the remaining male and female students:
remaining_descending_males_batch_records_indexed_dict_copy contains the remaining male students.
remaining_ascending_females_batch_records_indexed_dict_copy contains the remaining female students.
Finally, the function returns:
tutorial_groupings_dict_copy with the updated tutorial groups.
remaining_descending_males_batch_records_indexed_dict_copy with the remaining male students.
remaining_ascending_females_batch_records_indexed_dict_copy with the remaining female students.
tut_grp, the tutorial group for the batch.
school_inequality, the calculated school inequality.


```python
def insert_initial(batch,records_indexed_dict_copy,tutorial_groupings_dict_copy):
    starting_index=(batch-1)*50
    ending_index=batch*50
    tut_grp=records_indexed_dict_copy[starting_index]['Tutorial Group']
    batch_records_indexed_dict_copy = {k: v for k, v in records_indexed_dict_copy.items() if starting_index <= k < ending_index}
    schools={}
    for key, value in batch_records_indexed_dict_copy.items():
        # Extract the value for the "scholl" key in the nested dictionary
        school = value.get('School')
        if school in schools.keys():
            schools[school]=schools[school]+1
        else:
            schools[school]=1
    school_inequality=calculate_gini_coefficient_school_no_penalty(schools)
    males_batch_records_indexed_dict_copy = {k: v for k, v in batch_records_indexed_dict_copy.items() if v['Gender'] == 'Male'}
    females_batch_records_indexed_dict_copy = {k: v for k, v in batch_records_indexed_dict_copy.items() if v['Gender'] == 'Female'}
    descending_males_batch_records_indexed_dict_copy = sorter(males_batch_records_indexed_dict_copy)
    descending_males_batch_records_indexed_dict_copy = {index: value for index, (_, value) in enumerate(descending_males_batch_records_indexed_dict_copy.items())}
    ascending_females_batch_records_indexed_dict_copy = sorter(females_batch_records_indexed_dict_copy,descending = True)
    ascending_females_batch_records_indexed_dict_copy = {index: value for index, (_, value) in enumerate(ascending_females_batch_records_indexed_dict_copy.items())}
    for i in range(0,10):
        tutorial_groupings_dict_copy[tut_grp][i].append(descending_males_batch_records_indexed_dict_copy[i])
        tutorial_groupings_dict_copy[tut_grp][i].append(ascending_females_batch_records_indexed_dict_copy[i])
    remaining_descending_males_batch_records_indexed_dict_copy = {k: v for k, v in descending_males_batch_records_indexed_dict_copy.items() if k not in range(10)}
    remaining_ascending_females_batch_records_indexed_dict_copy = {k: v for k, v in ascending_females_batch_records_indexed_dict_copy.items() if k not in range(10)}
    return tutorial_groupings_dict_copy,remaining_descending_males_batch_records_indexed_dict_copy,remaining_ascending_females_batch_records_indexed_dict_copy,tut_grp,school_inequality
```

__13. Insert Rest__
We are now going to insert the rest of the students into the groups.
1. Merge the dictionaries of males and females so that the dataset will be in a single list 
2. Next, loop through the remaining students to check all available groups to see which group can accommodate the student without exceeding the size limit
If the size of a group is less than 5, we’ll do a score calculation using school_inequality to see which group is the best group to add a new student
i. Calculate the current score of the group without the new student 
ii. Simulate adding the current student into this group
iii. Calculate the new score after adding the student
iv. Compute the change in the score using the new score-old score 
v. Add the changes in score into the change list   
Else, if a group is already full, we add a large value (9999999999999999) to avoid selecting the group 
3. Print what the changes are and which group the student is added to 
4. Append the student to the selected group
5. Return the updated tutorial groupings

```python
def insert_rest(tut_grp,tutorial_groupings_dict_copy,males,females,school_inequality):
    remaining,amt=merge_dictionaries(males,females)
    for i in range(amt):
        #print(f"Right now we are at student {i+1}")
        student=remaining[i]
        change=[]
        for j in range(10):
            if len(tutorial_groupings_dict_copy[tut_grp][j])<5:
                old_score=calculate_stats(tutorial_groupings_dict_copy[tut_grp][j],school_inequality)
                new_students=tutorial_groupings_dict_copy[tut_grp][j].copy()
                new_students.append(student)
                new_score=calculate_stats(new_students,school_inequality)
                change_in_score=new_score-old_score
                change.append(change_in_score)
            else:
                change.append(9999999999999999)
        group=change.index(min(change))
        #print(f"The change is {change}")
        #print(f"We add this student to group {group}")
        tutorial_groupings_dict_copy[tut_grp][group].append(student)
    return tutorial_groupings_dict_copy
    
```

__14. Sort-Output Function:__

Now, we’ll need to sort the rest of the students for each tutorial group by looping through the batches and sorting them. 
1. We’ll create copies of tutorial_groupings_dict and records_indexed_dict so that we are able to work with it, preserving the original data 
2. Initialize batch counter to 1 before begin looping the tutorial groups 
Now we’ll loop through each batch; while batch is less than or equal to num_batch+1: 
i. We call insert_inital to assign initial groups for the current batch. This returns updated tutorial groups, remaining male and female students, the tutorial group identifier, and school inequality metric 
ii. We then allocate the remaining students to groups by calling insert_rest. This updates the tutorial groupings based on diversity constraints 
Print the completion status for the current batch 
Move on to the next batch 
3. Return the final tutorial groupings after all batches are processed. 


```python
def sort(num_batch,tutorial_groupings_dict,records_indexed_dict):
    tutorial_groupings_dict_copy=tutorial_groupings_dict
    records_indexed_dict_copy=records_indexed_dict
    batch=1
    while batch<num_batch+1:
        tutorial_groupings_dict_copy,remaining_batch_males,remaining_batch_females,tut_grp,school_inequality=insert_initial(batch,records_indexed_dict_copy,tutorial_groupings_dict_copy)
        tutorial_groupings_dict_copy=insert_rest(tut_grp,tutorial_groupings_dict_copy,remaining_batch_males,remaining_batch_females,school_inequality) 
        print(f"Batch {batch} is done for tutorial group {tut_grp}")
        batch+=1 
    return tutorial_groupings_dict

output=sort(120,tutorial_groupings_dict,records_indexed_dict)
print(output)
```

```python
check_group = "G-69"
len(output[check_group][0])
```

```python
def count_males_in_nested_lists(data):
    male_counts = []
    
    # Iterate over each nested list
    for nested_list in data:
        # Count dictionaries with "Gender" equal to "Male"
        male_count = sum(1 for d in nested_list if d.get('Gender') == 'Male')
        male_counts.append(male_count)
    
    return male_counts



result = count_males_in_nested_lists(output[check_group])
print(result)  # This will give you a list with the count of "Male" in each nested list
```

```python
from collections import Counter

def count_schools_in_nested_lists(data):
    school_counts_list = []
    
    # Iterate over each nested list
    for nested_list in data:
        # Extract the "School" values and count their occurrences
        school_counter = Counter(d.get('School') for d in nested_list if 'School' in d)
        school_counts_list.append(school_counter)
    
    return school_counts_list



result = count_schools_in_nested_lists(output[check_group])
for i, school_count in enumerate(result):
    print(f"List {i+1}: {school_count}")
```

```python
import math

def calculate_standard_deviation(data):
    # Function to calculate standard deviation
    n = len(data)
    mean = sum(data) / n
    variance = sum((x - mean) ** 2 for x in data) / n
    return math.sqrt(variance)

# List to store standard deviations
std_devs = []

# Calculate standard deviation for each of the 10 lists
for sublist in output[check_group]:
    cgpa_values = [float(d['CGPA']) for d in sublist]
    std_dev = calculate_standard_deviation(cgpa_values)
    std_devs.append(std_dev)

# Output the results
for index, std_dev in enumerate(std_devs, 1):
    print(f"Standard deviation of list {index}: {std_dev:.2f}")
```

__16. Export to new csv file__
To export the team-assigned student's information from a dictionary called "output" to a new csv file, 
1. We first create a csv_file named "students_team_assignment.csv".
2. The csv_file is opened in write mode to insert the column headers ("Tutorial Group", "Student ID", "School", "Name", "Gender", "CGPA", and "Team Assigned").
3. The output is sorted in ascending order of tutorial group number (by sorting the keys in the dictionary). Using for loop, each student, from each team, from each tutorial group is assigned a team_number starting from 1 before the student's details are formatted as a row and written to the csv file.

```python
# Define the CSV file name
csv_file = 'students_team_assignment.csv'

# Define the column headers
column_headers = ["Tutorial Group", "Student ID", "School", "Name", "Gender", "CGPA", "Team Assigned"]

# Open the CSV file for writing
with open(csv_file, mode='w') as file:
    # Write the column headers
    file.write(','.join(column_headers) + '\n')
    
    # Sort the tutorial groups using a custom key
    sorted_keys = sorted(output.keys(), key=lambda x: int(x.split('-')[1]))
    
    for tutorial_group in sorted_keys:
        teams = output[tutorial_group]
        # Loop through each team and keep track of the team number
        for team_number, team in enumerate(teams, start=1):
            # Loop through each student in the team
            for student in team:
                # Create a row with student details, including the team number
                row = [
                    tutorial_group,
                    student['Student ID'],
                    student['School'],
                    student['Name'],
                    student['Gender'],
                    student['CGPA'],
                    team_number
                ]
                # Write the row to the CSV file
                file.write(','.join(map(str, row)) + '\n')
```
